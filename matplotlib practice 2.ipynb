```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color='red',linestyle='--',marker='o')
plt.legend(['Line plotting'])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

```python
#x=['A','B','C','D','E']
#y=[10,5,8,3,7]
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.bar(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar plot')
plt.show()
```

```python
data=np.random.randn(5000)
plt.hist(data, bins=30,edgecolor='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

```python
data=np.random.randn(500)
plt.boxplot(data)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Box plot')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[5,8,2,10,4]
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[1,3,5,7,9]
plt.stackplot(x,y,z,labels=['X','Y','Z'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area plot')
plt.legend()
plt.show()
```

```python
data=np.random.randn(5,5)
plt.imshow(data,cmap='hot')
plt.colorbar()
plt.title('Heat map')
plt.show()
```

```python
theta=np.linspace(0,2*np.pi,100)
r=np.sin(3*theta)
plt.polar(theta,r)
plt.title('Polar plot')
plt.show()
```
