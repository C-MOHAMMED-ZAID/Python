```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.show()
```

```python
plt.bar(x,y)
plt.scatter(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('plot 1')
axes[1].scatter(x,y)
axes[1].set_title('plot 2')
plt.tight_layout()
plt.show()
```

```python
labels=['A','B','C','D']
values1=[10,15,7,12]
values2=[8,12,10,5]
x=range(len(labels))
width=0.35
fig,ax=plt.subplots()
rects1=ax.bar(x, values1, width, label='Group 1')
rects2=ax.bar([i+width for i in x],values2, width,label='Group 2')
ax.set_ylabel('Values')
ax.set_xlabel('Labels')
ax.set_title('Bar plot with multiple Bars')
ax.set_xticks([i+width/2 for i in x])
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
import numpy as np
data=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(data,bins=30)
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histogram')
plt.show()
```

```python
data=np.random.randn(10,10)
corr=np.corrcoef(data)
fig,ax=plt.subplots()
im=ax.imshow(corr,cmap='coolwarm')
cbar=ax.figure.colorbar(im,ax=ax)
ax.set_title('correlation matrix')
plt.show()
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2+Y**2))
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D surface plot')
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.plot(x,y1)
ax1.set_title('Sine')
ax2.plot(x,y2)
ax2.set_title('Cosine')
ax1.set_ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
fig,ax=plt.subplots()
ax.plot(x,y1,label='Sine')
ax.plot(x,y2,label='Cosine')
ax.legend()
ax.annotate('Maximum',xy=(np.pi/2,1),xytext=(np.pi/2,1.5),arrowprops=dict(facecolor='black',arrowstyle='->'))
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sine and Cosine functions')
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
fig,ax=plt.subplots()
ax.plot(x,y,color='red',linestyle='dashed',linewidth=2,marker='o',markersize=5,label="sine function")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Customized Sine Function')
ax.grid(True)
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.sin(2*x)
fig,ax=plt.subplots()
ax.plot(x,y1,label='Sine')
ax.plot(x,y2,label='Cosine')
ax.plot(x,y3,label='2*Sine')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Multiple line plots')
ax.legend()
plt.show()
```

```python
categories=['A','B','C','D']
sub_categories=['Sub1','Sub2','Sub3','Sub4']
data=[[10,15,7,12],
     [5,10,3,8],
     [8,12,10,5],
     [3,8,5,15]]
fig,ax=plt.subplots()
ax.bar(categories,data[0],label=sub_categories[0])
for i in range(1,len(sub_categories)):
    ax.bar(categories,data[i],bottom=np.sum(data[:i],axis=0),label=sub_categories[i])
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked bar chart')
ax.legend()
plt.show()
```

```python
theta=np.linspace(0,2*np.pi,100)
r=np.abs(np.sin(3*theta))
fig,ax=plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta,r)
ax.set_title('Polar plot')
plt.show()
```

```python
x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)
X,Y=np.meshgrid(x,y)
Z=np.exp(-X**2-Y**2)
fig,ax=plt.subplots()
contour=ax.contour(X,Y,Z)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Contour plot')
cbar=fig.colorbar(contour)
plt.show()
```

```python
np.random.seed(0)
n=100
x=np.random.rand(n)
y=np.random.rand(n)
z=np.random.rand(n)
colors=np.random.rand(n)
sizes=100*np.random.rand(n)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=colors,s=sizes,alpha=0.5)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter plot')
plt.show()
```

```python
x,y=np.meshgrid(np.linspace(-2,2,20),np.linspace(-2,2,20))
u=-y
v=x
fig,ax=plt.subplots()
ax.streamplot(x,y,u,v)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Stream plot')
plt.show()
```
